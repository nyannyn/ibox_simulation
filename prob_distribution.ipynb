### Ver.3
- 計算一個分局的郵箱全滿機率
- 直接計算一個時間單位(秒)的機率分布(省略填補0的過程，反正0不用加)

```python
import pandas as pd
import numpy as np

# 初始化所有需要的箱體和儲格
box_info = {
    1779: list(range(1, 41)),
    2621: list(range(1, 41)),
    2233: list(range(1, 41))
}

# 初始化86400秒的計數器，用來累計所有箱體和儲格每天的事件
event_counts = np.zeros(86400, dtype=int)

# 獲取所有日期範圍
all_dates = pd.to_datetime(status_filtered['資料建立日期時間']).dt.date.unique()

# 遍歷所有的箱體和儲格
for box_id, grid_ids in box_info.items():
    for grid_id in grid_ids:
        # 遍歷所有日期
        for date in all_dates:
            # 選取該箱體和儲格在當天的事件資料
            group = status_filtered[
                (status_filtered['箱體ID'] == box_id) &
                (status_filtered['儲格ID'] == grid_id) &
                (status_filtered['資料建立日期時間'].dt.date == date)
            ]

            if not group.empty:
                # 若有事件資料，計算每個事件間的秒數範圍
                events = list(zip(group['資料建立日期時間'], group['類別'].astype(int)))
                for i, (event_time, event_type) in enumerate(events):
                    # 轉換為一天中的秒數
                    seconds_since_midnight = event_time.hour * 3600 + event_time.minute * 60 + event_time.second
                    
                    # 計算寄件或取件的影響範圍
                    if event_type == 1:  # 寄件，狀態轉為 1
                        if i < len(events) - 1:
                            next_seconds_since_midnight = (
                                events[i + 1][0].hour * 3600 +
                                events[i + 1][0].minute * 60 +
                                events[i + 1][0].second
                            )
                        else:
                            next_seconds_since_midnight = 86400  # 到一天的結尾
                        
                        # 累加事件範圍內的秒數
                        event_counts[seconds_since_midnight:next_seconds_since_midnight] += 1

# 計算每秒的機率分布
# 總天數是 (所有箱體 x 所有儲格 x 所有日期)
total_days = len(box_info) * 40 * len(all_dates)
probability_distribution = event_counts / total_days

# 將機率分布轉換為 DataFrame 並儲存
probability_df = pd.DataFrame(probability_distribution, columns=['Probability'])
probability_df.index.name = 'Second'
probability_df.to_csv("每日秒級機率分布.csv", encoding="utf-8-sig")

print("每日秒級機率分布已儲存為 '每日秒級機率分布.csv'")
```

### Ver.4 計算Utilization Time 分布
- 一個儲格 放->取 所花費的時間 

```python

# 建立 DataFrame 並轉換時間格式
df = pd.DataFrame(status_filtered)
df['資料建立日期時間'] = pd.to_datetime(df['資料建立日期時間'])

# 排序資料（確保時間順序）
df = df.sort_values(['箱體ID', '儲格ID', '資料建立日期時間'])

# 計算時間差
time_deltas = []

# 分組並計算
for (box_id, cell_id), group in df.groupby(['箱體ID', '儲格ID']):
    # 找出 類別 = 1 到 0 的位置及時間
    group = group.reset_index(drop=True)
    for i in range(len(group) - 1):
        if group.loc[i, '類別'] == 1 and group.loc[i + 1, '類別'] == 0:
            time_diff = group.loc[i + 1, '資料建立日期時間'] - group.loc[i, '資料建立日期時間']
            time_deltas.append({
                '箱體ID': box_id,
                '儲格ID': cell_id,
                '從1到0時間差': time_diff
            })

# 將結果轉為 DataFrame 並顯示
result_df = pd.DataFrame(time_deltas)
result_df
```

```python
# 設置圖形大小
plt.figure(figsize=(10, 6))

# 繪製散點圖 (顛倒 x 和 y)
plt.scatter(
    x=result_df['從1到0時間差_天數'],
    y=result_df.index,  # 使用索引作為 y 軸，或改用其他可辨識的順序
    color='blue', alpha=0.6
)

# 設定圖形範圍，x 軸範圍設為 0 到 3 天
plt.xlim(0, 3)
plt.ylim(-1, len(result_df))  # 設定 y 軸範圍覆蓋所有資料點

# 標題和標籤
plt.title('指南郵局，儲格 放->取 所花費的時間  (0 - 3 天)')
plt.xlabel('從1到0時間差 (天)')
plt.ylabel('資料點編號')

# 顯示圖形
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from datetime import timedelta

# 假設 status_filtered 已存在並已經轉換為 DataFrame 格式
# 並且 time_diff_0_to_1_df 已經包含計算好的時間差

# 定義箱體和儲格資訊
box_info = {
    1779: list(range(1, 41)),
    2621: list(range(1, 41)),
    2233: list(range(1, 41))
}

# 檢查並補充缺失的儲格資料
all_data = []
for box_id, cells in box_info.items():
    for cell_id in cells:
        # 檢查 time_diff_0_to_1_df 中是否有該箱體和儲格的資料
        match = time_diff_0_to_1_df[(time_diff_0_to_1_df['箱體ID'] == box_id) & (time_diff_0_to_1_df['儲格ID'] == cell_id)]
        
        if not match.empty:
            # 如果有多筆資料，逐筆添加
            for _, row in match.iterrows():
                all_data.append({
                    '箱體ID': box_id,
                    '儲格ID': cell_id,
                    '從0到1時間差': row['從0到1時間差']
                })
        else:
            # 如果沒有資料，設置時間差為 365 days
            all_data.append({
                '箱體ID': box_id,
                '儲格ID': cell_id,
                '從0到1時間差': timedelta(days=365)
            })

# 最終結果 DataFrame
complete_time_diff_df = pd.DataFrame(all_data)
complete_time_diff_df

```
